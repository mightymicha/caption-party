```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
data = pd.read_csv("../rsc/data.csv")
```

```python
parties = ['linke', 'spd', 'grüne', 'cdu', 'fdp', 'afd']
colors = {'linke': '#FFC0CB',
          'grüne': '#42923bff',
          'spd': '#e2001aff',
          'cdu': '#252422ff',
          'fdp': '#ffec01ff',
          'afd': '#009ee0ff'}
```

```python
data.set_index('videoId', inplace=True)
```

```python
data.sort_values('viewCount')[['title', 'channelTitle','publishedAt','viewCount']]
```

```python
data[['title', 'channelTitle','publishedAt','viewCount', 'commentCount']].sort_values('commentCount')
```

```python
data['publishedAt'] = pd.to_datetime(data['publishedAt']) 
```

```python
viewCount = data.groupby([data["publishedAt"].dt.week, data['party']]).sum()['viewCount'].unstack()
viewCount
```

```python
count = data.groupby([data["publishedAt"].dt.week, data['party']])['title'].count().unstack()
count.plot(color=count.apply(lambda x: colors[x.name]))
```

```python
viewCount.plot(kind='bar', color=viewCount.apply(lambda x: colors[x.name]))
```
